# Création d'un échantillon de taille n
*Rappel du cours: un échantillon de taille n est la liste des résultats obtenus lorsqu’on répète n fois une même expérience aléatoire de façon indépendante.*

... C'est ce que nous avons fait dans le programme précédent : nous avons répété 10 fois une simulation de pile ou face avec une pièce équilibrée.

Cependant, il n'est pas facile d'établir la fréquence d'apparition du "1", il faut les compter à la main. Nous allons créer une fonction, nommée échantillon(n), qui renvoie la fréquence d'apparition du "1" dans un échantillon  de n simulations de pile ou face :  

```python
def échantillon(n):
    compteur=0          # le compteur sert à compter le nombre de "1". au début, il n'y en a pas
    for i in range (n):
        res=randint(0,1)
        #print(res)      #cette instruction est inutile, mais permet de calculer à la main le résultat.
        if res==1:
            compteur=compteur+1  #si le résultat est "1", on augment de le compteur...
    return(compteur/n)  # la fréquence est égale au quotient du nombre de succès ici de "1" sur le nombre total.
```

```python
échantillon(1000)
```
